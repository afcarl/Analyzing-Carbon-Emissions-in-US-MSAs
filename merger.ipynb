```python
__author__ = 'Shi Fan'
import csv
import pandas as pd
import numpy as np
import sys
```

```python
df_arcgis = pd.read_csv('data/ArcGIS_Census.csv')
df_carb = pd.read_csv('data/total_carbon_emission.csv')
df_arcgis.head()
```

```python
col_list = [5,6,28]
df_arc_cut = df_arcgis[col_list]
df_arc_cut.head()
```

```python
df_carb.columns
```

```python
df_carb = df_carb.rename(columns={' County': 'County', ' FIPS': 'FIPS', ' Total': 'Total'})
```

```python
df_merge = pd.merge(df_carb, df_arc_cut, how='inner', on='FIPS', sort=True)
```

```python
df_merge.head()
```

```python
# convert the unit from million tons to tons
for i in df_merge.columns:
    if (i=='ALAND10'):
        df_merge[i] /= 1000000
    elif (i!='State') & (i!='County') & (i!='FIPS') & (i!='POP2000'):
        df_merge[i] *= 1000000
df_merge.head()
```

```python
df_table = pd.read_csv('data/MSAtoCounty.csv')
df_table.head()
```

### As of February 2013, the Poughkeepsie-Newburgh-Middletown MSA no longer officially exists.[4] Dutchess and Orange Counties are now part of the New York-Newark-Jersey City, NY-NJ-PA Metropolitan Statistical Area; within this area, Orange County is a part of the New York-Jersey City-White Plains, NY-NJ Metropolitan Division, while Dutchess County is in the Dutchess County-Putnam County, NY Metropolitan Division.

```python
indexes = df_table[df_table['MSA Code']==39100].index
for i in indexes:
    df_table.loc[i,'MSA Code'] = 35620
```

```python
colMSA = [0,1,2]
df_tab_cut = df_table[colMSA]
df_tab_cut.head()
```

```python
df_final = pd.merge(df_merge, df_tab_cut, how='inner', on='FIPS', sort=True)
df_final.head()
```

```python
unicodes = df_final['MSA Code'].unique()
unicodes.sort()
#print unicodes
```

```python
df = pd.DataFrame(columns=df_final.columns[3:])
df['MSA Code'] = unicodes
for col in df.columns:
    if col!='MSA Code':
        df[col] = (df_final[col].groupby(df_final['MSA Code']).sum()).tolist()
df.head()
```

```python
df_norm = pd.DataFrame(columns=df.columns)
df_norm['POP2000'] = df['POP2000']
df_norm['MSA Code'] = df['MSA Code']
df_norm['ALAND10'] = df['ALAND10']
df_norm['MSA Name'] = df['MSA Name']
for col in df_norm.columns:
    if (col!='POP2000') & (col!='MSA Code') & (col!='ALAND10') & (col!='MSA Name'):
        df_norm[col] = df[col]/df['POP2000']
df_norm.head()
```

```python
df_norm.to_csv('data/per_capita_msa.csv')
```
